```python
%matplotlib nbagg
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import integrate
import numpy as np
def roundeffort(x):
    return 10*round(int(x)/10)
Abstract = pd.read_csv('/mnt/ceph/storage/data-tmp/2021/akpdk/AusgangslistenHiCal/AbstractTitelVolltext/Abstract/output/bmi_42.avg.precisionrecall', delimiter=r"\s+")
Abstract['approach'] = 'HiCal mit Abstract'
Titel = pd.read_csv('/mnt/ceph/storage/data-tmp/2021/akpdk/AusgangslistenHiCal/AbstractTitelVolltext/Titel/output/bmi_42.avg.precisionrecall', delimiter=r"\s+")
Titel['approach'] = 'HiCal mit Titel'
Volltext = pd.read_csv('/mnt/ceph/storage/data-tmp/2021/akpdk/AusgangslistenHiCal/AbstractTitelVolltext/Volltext/output/bmi_42.avg.precisionrecall', delimiter=r"\s+")
Volltext['approach'] = 'HiCal mit Volltext'
```

```python
df = pd.concat([Abstract,Titel,Volltext])
df['effort_rounded'] = df.effort.apply(lambda i: roundeffort(i))
sns.lineplot(data = df, x = 'effort_rounded', y = 'recall', hue = 'approach')
```
