## Recall von HiCal auf CEEDER_2018 mithilfe von den Strategien: Titel, Zitate, Referenzen

Zusätzlich noch mit Stemming und Stopping

```python
%matplotlib nbagg
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import integrate
import numpy as np
def roundeffort(x):
    return 10*round(int(x)/10)
StandardNormal = pd.read_csv('/mnt/ceph/storage/data-tmp/2021/akpdk/AusgangslistenHiCal/ceeder2018/Standard/Normal/output/bmi_42.avg.precisionrecall', delimiter=r"\s+")
StandardNormal['approach'] = 'HiCal'
StandardReferenzen = pd.read_csv('/mnt/ceph/storage/data-tmp/2021/akpdk/AusgangslistenHiCal/ceeder2018/Standard/Referenzen/output/bmi_42.avg.precisionrecall', delimiter=r"\s+")
StandardReferenzen['approach'] = 'HiCal mit DeepCT(Referenzen)'
StandardReferenzenStemStop = pd.read_csv('/mnt/ceph/storage/data-tmp/2021/akpdk/AusgangslistenHiCal/ceeder2018/Standard/ReferenzenStemStop/output/bmi_42.avg.precisionrecall', delimiter=r"\s+")
StandardReferenzenStemStop['approach'] = 'HiCal mit DeepCT(ReferenzenStemStop)'
StandardTitel = pd.read_csv('/mnt/ceph/storage/data-tmp/2021/akpdk/AusgangslistenHiCal/ceeder2018/Standard/Titel/output/bmi_42.avg.precisionrecall', delimiter=r"\s+")
StandardTitel['approach'] = 'HiCal mit DeepCT(Titel)'
StandardTitelStemStop = pd.read_csv('/mnt/ceph/storage/data-tmp/2021/akpdk/AusgangslistenHiCal/ceeder2018/Standard/TitelStemStop/output/bmi_42.avg.precisionrecall', delimiter=r"\s+")
StandardTitelStemStop['approach'] = 'HiCal mit DeepCT(TitelStemStop)'
StandardZitate = pd.read_csv('/mnt/ceph/storage/data-tmp/2021/akpdk/AusgangslistenHiCal/ceeder2018/Standard/Zitate/output/bmi_42.avg.precisionrecall', delimiter=r"\s+")
StandardZitate['approach'] = 'HiCal mit DeepCT(Zitate)'
StandardZitateStemStop = pd.read_csv('/mnt/ceph/storage/data-tmp/2021/akpdk/AusgangslistenHiCal/ceeder2018/Standard/ZitateStemStop/output/bmi_42.avg.precisionrecall', delimiter=r"\s+")
StandardZitateStemStop['approach'] = 'HiCal mit DeepCT(ZitateStemStop)'
```

```python
df = pd.concat([StandardNormal,StandardReferenzen,StandardZitate,StandardTitel])
df['effort_rounded'] = df.effort.apply(lambda i: roundeffort(i))
sns.lineplot(data = df, x = 'effort_rounded', y = 'recall', hue = 'approach')
```

```python
df = pd.concat([StandardNormal,StandardReferenzenStemStop,StandardZitateStemStop,StandardTitelStemStop])
df['effort_rounded'] = df.effort.apply(lambda i: roundeffort(i))
sns.lineplot(data = df, x = 'effort_rounded', y = 'recall', hue = 'approach')
```
