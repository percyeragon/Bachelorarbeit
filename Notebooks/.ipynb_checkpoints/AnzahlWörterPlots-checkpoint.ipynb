```python
import json
import pandas as pd
import seaborn as sns
```

```python
liste=[]
volltextListe = []
with open("/mnt/ceph/storage/data-tmp/2021/akpdk/AusgangslistenHiCal/AbstractTitelVolltext/Abstract/Ausgangsliste.json", 'r') as infile:
    data = json.load(infile)
    for obj in data:
        liste.append(["Abstract",obj["nummer"],len(obj["abstract"].split())])
with open("/mnt/ceph/storage/data-tmp/2021/akpdk/AusgangslistenHiCal/AbstractTitelVolltext/Titel/Ausgangsliste.json", 'r') as infile:
    data = json.load(infile)
    for obj in data:
        liste.append(["Titel",obj["nummer"],len(obj["titel"].split())])
with open("/mnt/ceph/storage/data-tmp/2021/akpdk/AusgangslistenHiCal/AbstractTitelVolltext/Volltext/Ausgangsliste.json", 'r') as infile:
    data = json.load(infile)
    for obj in data:
        volltextListe.append(["Volltext",obj["nummer"],len(obj["abstract"].split())])
df = pd.DataFrame(liste,columns=['Feld', 'docno','anzahlTerme'])
dfVolltext = pd.DataFrame(volltextListe,columns=['Feld', 'docno','anzahlTerme'])
```

```python
g=sns.displot(data = df, x = 'anzahlTerme', hue = 'Feld')
g.set(xlim=(1, 400))
```

```python
g=sns.displot(data = dfVolltext, x = 'anzahlTerme', hue = 'Feld')
g.set(xlim=(1, 14000))
```

```python
df = df.append(dfVolltext)
g=sns.barplot(data=df,y="anzahlTerme", x="Feld",hue="Feld")
```
